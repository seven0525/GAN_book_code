```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import glob
import os
```

# 10%マージン

```python
def mirror_padding(img_path):
    img1 = cv2.imread(img_path)
    padding_y = img1.shape[0] // 10
    padding_x = img1.shape[1] // 10
    img2 = cv2.copyMakeBorder(img1, padding_y, padding_y, padding_x, padding_x, cv2.BORDER_REFLECT_101)
    return img2    

image_paths = glob.glob('../../anime/*')
for image_path in image_paths:
    img_name = image_path.split('/')[-1]
    img = mirror_padding(image_path)
    cv2.imwrite('./anime_output/' + img_name, img)
```

# OK画像とNG画像の仕分け

```python
import shutil

files = glob.glob("./anime_output/*")
f = open('./OK_10.txt')
OK_files = f.read()
f.close()
OK_files = OK_files.split(",")
OK_files = list(set(OK_files))

for OK_file in OK_files:
    path = "./anime_output/" + OK_file
    if os.path.exists(path):
        shutil.move(path, './anime_face_OK_10/')
        
files = glob.glob("./anime_output/*")
f = open('./NG_10.txt')
NG_files = f.read()
f.close()
NG_files = NG_files.split(",")
NG_files = list(set(NG_files))

for NG_file in NG_files:
    path = "./anime_output/" + NG_file
    if os.path.exists(path):
        shutil.move(path, './anime_face_NG/')
```
